```python
import os
import random
from shutil import copyfile, copytree
from tqdm import tqdm
```

```python
isl_folder_path = '/content/drive/MyDrive/sign-language/isl'
isl_small_folder_path = '/content/drive/My Drive/isl_small'

# Create ISL_small directory if not exists
if not os.path.exists(isl_small_folder_path):
    os.makedirs(isl_small_folder_path)

# Function to randomly select 70 images from a folder
def select_images(folder_path):
    images = os.listdir(folder_path)
    selected_images = random.sample(images, 70)
    return selected_images

# Loop through each subfolder in ISL folder
for subfolder_name in tqdm(os.listdir(isl_folder_path)):
    subfolder_path = os.path.join(isl_folder_path, subfolder_name)
    if os.path.isdir(subfolder_path):
        # Create corresponding subfolder in ISL_small
        isl_small_subfolder_path = os.path.join(isl_small_folder_path, subfolder_name)
        if not os.path.exists(isl_small_subfolder_path):
            os.makedirs(isl_small_subfolder_path)

        # Select 70 random images from the subfolder
        selected_images = select_images(subfolder_path)

        # Copy selected images to ISL_small subfolder
        for image_name in selected_images:
            image_path = os.path.join(subfolder_path, image_name)
            destination_path = os.path.join(isl_small_subfolder_path, image_name)
            copyfile(image_path, destination_path)

print("Images copied successfully!")
```
