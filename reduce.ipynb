## Import necessary libraries

```python
import os
import random
from shutil import copyfile
from tqdm import tqdm
```

## Trim the number of samples in each class in the ISL dataset

Please add appropriate folder paths for the variables `islFolderPath` and `islSmallFolderPath`

```python
islFolderPath = '/path/to/original/dataset/'
islSmallFolderPath = '/path/to/trimmed/dataset/' # folder need not exist

# Create ISL_small directory if not exists
if not os.path.exists(islSmallFolderPath):
    os.makedirs(islSmallFolderPath)

# Function to randomly select 70 images from a folder
def select_images(folderPath):
    images = os.listdir(folderPath)
    selectedImages = random.sample(images, 70)
    return selectedImages

# Loop through each subfolder in ISL folder
for subfolderName in tqdm(os.listdir(islFolderPath)):
    subfolderPath = os.path.join(islFolderPath, subfolderName)
    if os.path.isdir(subfolderPath):
        # Create corresponding subfolder in ISL_small
        islSmallSubfolderPath = os.path.join(islSmallFolderPath, subfolderName)
        if not os.path.exists(islSmallSubfolderPath):
            os.makedirs(islSmallSubfolderPath)

        # Select 70 random images from the subfolder
        selectedImages = select_images(subfolderPath)

        # Copy selected images to ISL_small subfolder
        for image_name in selectedImages:
            image_path = os.path.join(subfolderPath, image_name)
            destination_path = os.path.join(islSmallSubfolderPath, image_name)
            copyfile(image_path, destination_path)

print("Images copied successfully...")
```

Copy `class 0` from ASL dataset to ISL dataset

```python
sourceFolder = "/path/to/asl/class/0"
destinationFolder = "/path/to/trimmed/isl/class/0" # folder need not exist

# Create the destination folder if it doesn't exist
if not os.path.exists(destinationFolder):
    os.makedirs(destinationFolder)

# Get the list of files in the source folder
files = os.listdir(sourceFolder)

# Iterate through the files and copy them to the destination folder
for filename in files:
      sourceFile = os.path.join(sourceFolder, filename)
      destinationFile = os.path.join(destinationFolder, filename)
      # Copy the file
      copyfile(sourceFile, destinationFile)
      print("Copied {} to {}".format(filename, destinationFile))

print("Copy process completed.")
```
